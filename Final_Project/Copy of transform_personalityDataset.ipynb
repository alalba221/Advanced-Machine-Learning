```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/CPSC_8420_Project/newPersonalityDataset2.csv")

df.head()
```

```python
letter_arr = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z'];

df_new = pd.concat([df, pd.DataFrame(columns=letter_arr)], axis=1)

df_new.fillna(0, inplace=True)

df_new[letter_arr] = df_new[letter_arr].astype(float)
```

```python
def letter_counter(df_new, row):

  j = 0

  for char in df_new.loc[row, 'posts']:

    i=0
    j=j+1

    for letter in letter_arr:

      i = i + 1

      if char == letter:

        df_new.iloc[row, i+6] = df_new.iloc[row, i+6] + 1

  i = 7

  while i<33:

    df_new.iloc[row, i] = df_new.iloc[row,i]/j

    i = i + 1

  return df_new
```

```python
row_num = 0

while row_num < (df_new.shape[0] - 5000):

  df_new = letter_counter(df_new, row_num)

  row_num = row_num + 1

df_new.head()
```

```python
df_new.to_csv('/content/drive/My Drive/CPSC_8420_Project/personality_letter1.csv', index=False)
```
